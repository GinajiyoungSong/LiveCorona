```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import urllib.request

fullurl = "https://namu.wiki/w/%EC%BD%94%EB%A1%9C%EB%82%98%EB%B0%94%EC%9D%B4%EB%9F%AC%EC%8A%A4%EA%B0%90%EC%97%BC%EC%A6%9D-19?from=%EC%BD%94%EB%A1%9C%EB%82%98%2019#fn-21"
req = urllib.request.Request(fullurl, headers={'User-Agent' : 'Mozilla/5.0'})
response=urllib.request.urlopen(req).read()
text = response.decode('utf-8')
soup = BeautifulSoup(text, 'lxml')
```

```python
peo = []
b=[]
```

```python
tables = soup.find_all("div",class_ = "wiki-table-wrap")[9]
trs = tables.find_all('tr')
a = trs[0]
tds = a.find_all('td')
for i in tds : 
    b1=i.get_text().strip()
    b.append(b1)
peo.append(b)
print(peo)
b=[]
```
